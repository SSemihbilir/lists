```python
# 1-  "Samsung S5,Samsung S6,Samsung S7,Samsung S8" elemanlarına sahip bir liste oluşturunuz.

telefonlar = ["Samsung S5","Samsung S6","Samsung S7","Samsung S8"]

# 2-  Liste Kaç elemanlıdır ?
print(len(telefonlar))                      #4

# 3-  Listenin ilk ve son elemanı nedir ?
print (telefonlar[0])                       #"Samsung S5”
print (telefonlar[-1])                      #"Samsung S8”

# 4-  "Samsung S5" değerini "Samsung S9" ile değiştirin.
telefonlar[0] = "Samsung S9"
print (telefonlar)                          #["Samsung S9","Samsung S6","Samsung S7","Samsung S8"]


# 5-  "Samsung S6" listenin bir elemanı mıdır ?
telefonlar = ["Samsung S5","Samsung S6","Samsung S7","Samsung S8"]

if "Samsung S6" in telefonlar:
      print("Samsung S6 liste içinde var.") #"Samsung S6 liste içinde var."

# 6-  Listenin -3 indeksindeki değer nedir ?
print (telefonlar[-3])                      #"Samsung S6"

# 7-  Listenin ilk 2 elemanını alın.

print(telefonlar[:2])                       #"Samsung S5","Samsung S6"

# 8-  Listenin son 2 elemanı yerine "Samsung S9" ve "Samsung S10" değerlerini ekleyin.

telefonlar[-2:] = ["Samsung S9","Samsung S10"]
print(telefonlar)                           #["Samsung S5","Samsung S6","Samsung S9","Samsung S10"]

# 9-  Listenin üzerine "IPhone X" ve "IPhone XR" değerlerini ekleyin.
sonuc  = telefonlar+["IPhone X", "IPhone XR"]
print(sonuc)                         #['Samsung S5', 'Samsung S6', 'Samsung S7', 'Samsung S8', 'IPhone X', 'IPhone XR']

# 10- Listenin son elemanını silin.
del telefonlar[-1]
print(telefonlar)                           #['Samsung S5', 'Samsung S6', 'Samsung S9', 'Samsung S10', 'IPhone X']

# 11- Liste elemanlarını tersten yazdırınız.
print(telefonlar[::-1])                     #['IPhone X', 'Samsung S10', 'Samsung S9', 'Samsung S6', 'Samsung S5']

# 12- Liste elemanlarını ekrana yazdırınız.
for a in telefonlar:
    print(a,end=" ")                        #Samsung S5 Samsung S6 Samsung S9 Samsung S10 IPhone X 
```

```python
# 1- Asagidaki oyuncu listesine uygun olacak sekilde kullanıcıdan oyuncu bilgilerini isteyiniz.

players = {
    '1': 
        {
            'name': 'Cristiano Ronaldo', 
            'yearOfBirth': '1985', 
            'nationality': 'Portugal', 
            'current_team': 'Portugal', 
            'history': ['Juventus', 'Real Madrid', 'Portugal']
        }, 
    '2': 
        {
            'name': 'Lionel Messi', 
            'yearOfBirth': '1987', 
            'nationality': 'Argentina', 
            'current_team': 'Barcelona', 
            'history': ['Barcelona', 'Argentina', 'Portugal']}
      }
# -------------------------------

a =int(input("players listesine eklenecek oyuncu sayısını giriniz:"))
b=0

while b<a:

    sd = input('oyuncu id: ')
    name = input('oyuncu ad: ')
    nationality = input("ülke: ")
    yearOfBirth = input('doğum yılı: ')
    current_team = input('takım: ')
    history = input('oynadığı yerler: ')

    players.update({
        sd: {
            "name": name,
            "yearOfBirth": yearOfBirth,
            "nationality": nationality,
            "yearOfBirth": yearOfBirth,
            "current_team": current_team,
            "history": history.split(',')
        }
    })
    b+=1

# 2- id' e göre arama yapınız.
sd = input('aramak istediğiniz oyuncu id: ')
player = players.get(sd)       #id si verilen oyuncunun bilgileri ekrana yazdırılır.

print(player)
print(f'name: {player.get("name")}')  #aynı oyuncunun ismi ekrana yazdırılır.  -->name: Cristiano Ronaldo

# 3- id' e göre bilgi kayıt siliniz.

sd = input('silmek istediğiniz oyun id: ')
players.pop(sd)         # id numarası verilen oyuncu silinir.

print(players)
```

```python
opelObj = {
    "marka": "Opel",
    "model": "Corsa",
    "yil": 2020
}

sonuc = opelObj["marka"]
print(sonuc)                    # Opel
sonuc = opelObj.get("marka")    # marka karşılıgına gelen Opel i yazdırır.
print(sonuc)                    # Opel

for x in opelObj:
    print(x)                    # marka - model - yil

for x in opelObj:
    print(opelObj[x])           # Opel - Corsa - 2020

for x in opelObj.values():      # valu bilgilerini yazdırmak için kullanılar.
    print(x)                    # Opel - Corsa - 2020

for x,y in opelObj.items():
    print(x,y)                  # marka Opel - model Corsa - yil 2020

sonuc = "marka" in opelObj      # opelObj icerisindeki key bilgileri arasında marka var mı sorgular
print(sonuc)                    # True

sonuc = len(opelObj)            # kac key girildiğini verir.
print(sonuc)                    # 3

opelObj["renk"] = "kırmızı"    # renk keyine kırmızı valu degerini atar.
print(opelObj)                 #{'marka': 'Opel', 'model': 'Corsa', 'yil': 2020, 'renk': 'kırmızı'}

opelObj.pop("renk")            # renk keyini siler.
print(opelObj)                 # {'marka': 'Opel', 'model': 'Corsa', 'yil': 2020}

opelObj.popitem()              #dictionarynin son keyini (yil) siler
print(opelObj)                 #{'marka': 'Opel', 'model': 'Corsa'}

del opelObj["marka"]           # marka keyini siler.
print(opelObj)                 # {'model': 'Corsa'}

opelObj = {'marka': 'Opel', 'model': 'Corsa', 'yil': 2020, 'renk': 'kırmızı'}

obj = opelObj.copy() # referans

obj["marka"] = "Mazda"
print(1,obj)                 #{'marka': 'Mazda', 'model': 'Corsa', 'yil': 2020, 'renk': 'kırmızı'}
print(2,opelObj)             #{'marka': 'Opel', 'model': 'Corsa', 'yil': 2020, 'renk': 'kırmızı'}

opelObj.update({"marka": "Bmw","renk": "Kırmızı"})
print(opelObj)              # {'marka': 'Bmw', 'model': 'Corsa', 'yil': 2020, 'renk': 'Kırmızı'}
print(obj)                  # {'marka': 'Mazda', 'model': 'Corsa', 'yil': 2020, 'renk': 'kırmızı'}

opelObj.clear()              # icerisini bosaltırız
print(opelObj)               # {} 

del opelObj                  # opelObj i yamamen sileriz.
#print(opelObj)              # silindigi icin hata alırız.
```

```python
_list = [1,2,3]
_tuple = 1,"iki",True
_tuple2 = (3,"dört",True)

print(type(_list))            # <class 'list'>
print(type(_tuple))           # <class 'tuple'>

print(_list[1])               # 2
print(_tuple[1])              # iki

print(len(_list))
print(len(_tuple))

_list[0]=5
# _tuple[0]=5                  #'tuple' nesnesi, öğe atamasını desteklemiyor

_list.append(3)
# _tuple.append(5)             # 'tuple' nesnesinin 'ekleme' özelliği yok

print(_tuple.count('iki'))     # 1 tane

print(_tuple + _tuple2)        # (1, 'iki', True, 3, 'dört', True)

_t = tuple([3,4,5])
print(type(_t))                # <class 'tuple'>
t = tuple(_list)
print(type(t))                 # <class 'tuple'>
print(_t,t)                    # (3, 4, 5) (5, 2, 3, 3)
```

```python
def maxsum(arr):    #bir listede art arda gelen sayıların max toplamını bulan fonk.
    basla = arr[0]
    toplam = 0
    for i in range(len(arr)):
        toplam += arr[i]
        if toplam <0:
            toplam = 0
        elif basla < toplam:
            basla = toplam
    return basla
```

# ENUMARATE AND ZİP

```python
markalar = "opel","bmw","mercedes"

#enumarate

print(type(enumerate(markalar)))             # <class 'enumerate'>
print(list(enumerate(markalar)))             # [(0, 'opel'), (1, 'bmw'), (2, 'mercedes')]

for index,value in enumerate(markalar,10):   # liste tuple (markalar), numaralamaya başlayacağı değer (10)
    print(f"{index}-{value}")                # 10-opel   # 12-bmw    # 13-mercedes
```

```python
# zip

liste1 = [1,2,3,4,5]
liste2 = ['a','b','c','d','e','f']
liste3 = [100,200,300,400,500]

zip(liste1,liste2,liste3)     # <zip at 0x20a2d168d00> list tuple set  cevrilince okunur
 
print(list(zip(liste1,liste2,liste3)))  # esleşemeyen degerler birleştirilmez.(f gibi)
                          # [(1, 'a', 100), (2, 'b', 200), (3, 'c', 300), (4, 'd', 400), (5, 'e', 500)]

for item in zip(liste1,liste2):
    print(item)           # (1, 'a') # (2, 'b') (3, 'c') # (4, 'd')  # (5, 'e')
    
for a,b,c in zip(liste1,liste2,liste3):
    print(a,b,c)          # 1 a 100  # 2 b 200  #  3 c 300  # 4 d 400  #  5 e 500
```
